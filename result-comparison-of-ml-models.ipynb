```python
import matplotlib.pyplot as plt

# Model names
models = ["Decision Tree", "KNN", "SVM", "AdaBoost", "XGBoost", "Logistic Regression", "Random Forest"]

# Accuracy values after upsampling
accuracies = [87.21, 70.34, 82.54, 59.20, 90.07, 69.62, 88.40]

# Plot the bar chart
plt.figure(figsize=(10, 5))
plt.bar(models, accuracies)

# Labels and title
plt.xlabel("Machine Learning Models")
plt.ylabel("Accuracy (%)")
plt.title("Comparison of Model Accuracy After Upsampling")
plt.ylim(0, 100)  # Accuracy ranges from 0 to 1
plt.xticks(rotation=45)

# Show values on top of bars
for i, acc in enumerate(accuracies):
    plt.text(i, acc + 0.02, f"{acc:.2f}", ha='center')

plt.show()
```
